<h1>Detect faces using Haarcascade Classifier in OpenCV</h1>

<h3>Modules's Importation</h3>

```python
import cv2
import os
import matplotlib.pyplot as plt
```

<h3>Initialising the Model </h3>

<p>Make to down the <strong style="color="red"">haarcascade_frontalface_default.xml</strong></p>

```python
classifier = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

<h3>Processing an  image for the face detection</h3>

```python
img = cv2.imread("me.jpg")
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
gray.shape
```

```python
plt.figure(figsize=[5,5])
plt.imshow(img)
plt.title("My profile Picture")
plt.show()
```

<h3>Predicting the image meaning output [face_or_not_present,x_anchor,y_anchor,width_box,height_box]</h3>

```python
faces = classifier.detectMultiScale(gray)
```

<h3>Drawing the shape according to outputs coordinates</h3>

```python
for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),1)
```

<h3>Plotting the output image with the box</h3>

```python
plt.imshow(img)
plt.text(70,20,"Face")
plt.show()
```

Here we go, we are done creating a very basic face detection using opencv,
